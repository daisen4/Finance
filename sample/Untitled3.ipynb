```python
'''
元ネタ：http://pineplanter.moo.jp/non-it-salaryman/2018/01/03/python-scraping/
'''
from bs4 import BeautifulSoup
import urllib.request as req
import numpy as np
import time
import pandas as pd


page_num = 1000
stock_temp = []
for i in range(page_num):

    url="https://info.finance.yahoo.co.jp/history/?code=USDJPY%3DX&sy=2007&sm=8&sd=20&ey=2018&em=11&ed=18&tm=d&p=" + str(i+1)
    print (url)
    res = req.urlopen(url)
    soup = BeautifulSoup(res, 'html.parser');
    #print(soup)
    stock_extract = [value.get_text() for value in soup.find_all("td")[2:102]]
    stock_temp.extend(stock_extract)
    
    time.sleep(0.5)
    
stock_temp = np.array(stock_temp)

stock = stock_temp.reshape(int(len(stock_temp)/5), 5)
#print(stock[:,1:5])
stock = pd.DataFrame(stock[:,1:5], columns=["start", "high", "low", "end"], index=stock[:,0])

# 株価のカラムが文字列になっていて、かつカンマが入っているのでカンマを除去してfloat型にする。
#for i in range(4):
    #stock.ix[:,i] = stock.ix[:,i].str.replace(",", "").astype(float)
    #stock[:,i] = stock[:,i].astype(float)

print(stock)
#values = soup.select_one("#USDJPY_detail_bid").findAll(text=True)
#current_value = float(''.join(values))
#print(current_value)
```

```python
import urllib.request
import urllib.parse
import json
from pprint import pprint

# YQLをAPI経由で利用します
url = "https://info.finance.yahoo.co.jp/history/?code=USDJPY%3DX&sy=2007&sm=8&sd=20&ey=2018&em=11&ed=18&tm=d"
params = {
    "q": 'select * from yahoo.finance.xchange where pair in ("USDJPY")',
    "format": "json",
    "env": "store://datatables.org/alltableswithkeys"
}
url += "?" + urllib.parse.urlencode(params)
res = urllib.request.urlopen(url)

# 結果はJSON形式で受け取ることができます
result = json.loads(res.read().decode('utf-8'))
pprint(result)
"""
{'query': {'count': 1,
           'created': '2016-08-22T02:57:07Z',
           'lang': 'en-US',
           'results': {'rate': {'Ask': '100.6850',
                                'Bid': '100.6380',
                                'Date': '8/21/2016',
                                'Name': 'USD/JPY',
                                'Rate': '100.6380',
                                'Time': '10:58pm',
                                'id': 'USDJPY'}}}}
"""

# その中から必要な情報（今回はUSD→JPYの為替レート）を取得します
rate = result["query"]["results"]["rate"]["Rate"]
print('USD/JPY:', rate)
# USD/JPY: 100.6380
```
